```
plt.title("Comportamento da Multiplicidade de Estados")
plt.xlabel("z")
plt.ylabel("Multiplicidade")
x = np.linspace(0,1,100)
for i in [1,10,100,1000] : 
  y = (4*x*(1-x))**i
  plt.plot(x,y, label = "N=" + str(i))
plt.legend()
```

```
plt.title("Comportamento da Magnetização ")
plt.xlabel("b")
plt.ylabel("Magnetização")
b = np.linspace(0.005,3,100)
for j in [.5,1.,1.5,2.0] : 
  y = (j + .5)/np.tanh(b*(j + .5)) -.5/np.tanh(b*.5)
  plt.plot(b,y, label = "j=" + str(j))
plt.legend()
```

```
import numpy as np
import matplotlib.pyplot as plt

def ljp(r, epsilon, sigma):
    
    return 48 * epsilon * np.power(sigma, 12) / np.power(r, 13) \
    - 24 * epsilon * np.power(sigma, 6) / np.power(r, 7)
    
r = np.linspace(3.5, 8, 100)
plt.plot(r, ljp(r, 0.0103, 3.3))
plt.xlabel('distance')
plt.ylabel('energy')
plt.show()
```
